```python
"""
el_toro = wood_coasters[wood_coasters['Name'] == 'El Toro']
print(el_toro)

el_toro_ranking = el_toro['Rank'][el_toro['Park'] == 'Six Flags Great Adventure']
print(el_toro_ranking)

year_of_rank = el_toro['Year of Rank'][el_toro['Park'] == 'Six Flags Great Adventure']
print(year_of_rank)

ax = plt.subplot()
plt.plot(range(len(year_of_rank)), el_toro_ranking)
plt.xlabel("Year")
plt.ylabel("Ranking")
plt.title("El Toro Rankings")

ax.set_xticks(range(len(year_of_rank)))
ax.set_xticklabels(year_of_rank)
ax.set_yticks(el_toro_ranking)
ax.invert_yaxis()
plt.show()
"""

def coaster_ranking_over_time(name_of_coaster, park_name, material):
    if material == 'wood':
        park_coaster_name = wood_coasters[(wood_coasters['Name'] == name_of_coaster) & (wood_coasters['Park'] == park_name)]
    
    if material == 'steel':
        park_coaster_name = steel_coasters[(steel_coasters['Name'] == name_of_coaster) & (steel_coasters['Park'] == park_name)]
    
    
    x = park_coaster_name['Year of Rank']
    y = park_coaster_name['Rank']

    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    plt.plot(range(len(x)), y, marker = 'o')
    
    ax.set_xticks(range(len(x)))
    ax.set_xticklabels(x)
    ax.set_yticks(y)
    ax.invert_yaxis()
    
    plt.title(name_of_coaster + " Rankings")
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    plt.legend([name_of_coaster])
    
    plt.show()

coaster_ranking_over_time("El Toro", "Six Flags Great Adventure", 'wood')
```

```python
def two_coaster_rankings_over_time(coaster_one, coaster_one_park, coaster_one_material, coaster_two, coaster_two_park, coaster_two_material):
    
    if coaster_one_material == 'wood':
        park_coaster_name_one = wood_coasters[(wood_coasters['Name'] == coaster_one) & (wood_coasters['Park'] == coaster_one_park)]
    elif coaster_one_material == 'steel':
        park_coaster_name_one = steel_coasters[(steel_coasters['Name'] == coaster_one) & (steel_coasters['Park'] == coaster_one_park)]
        
    if coaster_two_material == 'wood':
        park_coaster_name_two = wood_coasters[(wood_coasters['Name'] == coaster_two) & (wood_coasters['Park'] == coaster_two_park)]
    elif coaster_two_material == 'steel':
        park_coaster_name_two = steel_coasters[(steel_coasters['Name'] == coaster_two) & (steel_coasters['Park'] == coaster_two_park)]
        
    
    x1 = park_coaster_name_one['Year of Rank']
    x2 = park_coaster_name_two['Year of Rank']
    
    y1 = park_coaster_name_one['Rank']
    y2 = park_coaster_name_two['Rank']
    
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    plt.plot(range(len(x1)), y1, marker = 'o')
    plt.plot(range(len(x2)), y2, marker = 'o')
    
    ax.set_xticks(range(len(x1)))
    ax.set_xticklabels(x1)
    ax.set_yticks(range(4 + 1))
    ax.invert_yaxis()
    
    plt.title(coaster_one + " vs. " + coaster_two + " Rankings")
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    plt.legend([coaster_one, coaster_two])
    
    plt.show()

two_coaster_rankings_over_time("El Toro", "Six Flags Great Adventure", "wood", "Boulder Dash", "Lake Compounce", "wood")
```

```python
def top_ten_rankings(num, material):
    if material == 'wood':
        top_num_ranking = wood_coasters[wood_coasters['Rank'] <= num]
    if material == 'steel':
        top_num_ranking = steel_coasters[steel_coasters['Rank'] <= num]
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    ax.set_yticks(range(num + 1))
    ax.invert_yaxis()
    
    for coaster in set(top_num_ranking['Name']):
        coaster_ranking = top_num_ranking[top_num_ranking['Name'] == coaster]
        plt.plot(coaster_ranking['Year of Rank'], coaster_ranking['Rank'], label = coaster, marker = 'o')
    
    plt.title("Top " + str(num) + " Rankings for Roller Coasters")
    plt.xlabel("Year")
    plt.ylabel("Ranking")
    
    plt.legend()
    plt.show()
    
top_ten_rankings(5, 'wood')
```

```python
roller_coasters = pd.read_csv('roller_coasters.csv')
roller_coasters.head()
```

```python
#speed = roller_coasters['speed']
#speed = speed.dropna()
#print(speed)

#height = roller_coasters['height']
#height = height.dropna()
#print(height)

def roller_coaster_speed(numeric_column_name, dataframe):
    dataframe = dataframe.dropna(subset = [numeric_column_name])
    column = dataframe[numeric_column_name].values
    
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    
    if numeric_column_name == 'height':
        plt.hist(column, range = (0, 140), edgecolor = 'gray')
    else:
        plt.hist(column, edgecolor = 'gray')
    
    plt.title("Histogram of Roller Coaster: " + numeric_column_name)
    plt.xlabel(numeric_column_name)
    plt.ylabel("Count")
    
    plt.show()
    
    
roller_coaster_speed('speed', roller_coasters)
roller_coaster_speed('height', roller_coasters)
roller_coaster_speed('length', roller_coasters)
roller_coaster_speed('num_inversions', roller_coasters)
```

```python
#num_inver = roller_coasters['num_inversions']
#print(num_inver)

#park = roller_coasters[roller_coasters['park'] == 'Six Flags Great Adventure']
#park = park.sort_values('num_inversions', ascending = False) => This sorts the the park table based on 'num_inversions'
#print(park)

#roller_coaster_by_park = roller_coasters.name[roller_coasters.park == 'Six Flags Great Adventure']
#print(roller_coaster_by_park)

def num_inversions_by_park(park, dataframe):
    park_coasters = dataframe[dataframe.park == park]
    park_coasters = park_coasters.sort_values('num_inversions', ascending = False)
    
    inver_by_park = park_coasters.num_inversions
    coaster_by_park = park_coasters.name
    
    x = coaster_by_park
    y = inver_by_park
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    plt.bar(range(len(x)), y)
    
    plt.title("Number of Inversions Per Coaster at " + park)
    plt.xlabel("Roller Coaster Name")
    plt.ylabel("# of Inversions")
    
    ax.set_xticks(range(len(x)))
    ax.set_xticklabels(x, rotation = 90)
    plt.show()
    
num_inversions_by_park('Six Flags Great Adventure', roller_coasters)
```

```python
#status = roller_coasters['status']
#print(status)

#status_open = roller_coasters[roller_coasters.status == 'status.operating']
#print(status_open)

def status_of_coaster(dataframe):
    
    status = []
    
    operating = dataframe[dataframe.status == 'status.operating']
    closed = dataframe[dataframe.status == 'status.closed.definitely']
    
    status.append(len(operating))
    status.append(len(closed))
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    plt.pie(status, autopct = '%0.1f%%', labels = ['Operating', 'Closed'], colors = ['lightblue', 'pink'])
    
    plt.title("Operation Status of Roller Coasters")
    
    plt.show()
    
status_of_coaster(roller_coasters)
```

```python
def quantitative_scatter_plot(column_one, column_two, dataframe):
    dataframe = dataframe.dropna()
    column_1 = dataframe[column_one].values
    column_2 = dataframe[column_two].values
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    
    if (column_one == 'height') | (column_two == 'height'):
        plt.scatter(column_1, column_2)
        plt.axis([0, 140, 0, 140])
    else:
        plt.scatter(column_1, column_2)
    
    plt.title("Scatter Plot: " + column_one + " vs. " + column_two)
    plt.xlabel(column_one)
    plt.ylabel(column_two)
    plt.show()
    
quantitative_scatter_plot('length', 'speed', roller_coasters)
quantitative_scatter_plot('height', 'speed', roller_coasters)
quantitative_scatter_plot('speed', 'num_inversions', roller_coasters)
```

```python
#seating_type = roller_coasters.seating_type
#print(seating_type)

#seating_type_count = roller_coasters.groupby('seating_type').name.count().reset_index()
#print(seating_type_count)

#seating = seating_type_count['seating_type'].values
#counts = seating_type_count['name'].values
#print(seating)
#print(counts)

def seating_type(dataframe):
    
    def addlabels(x, y):
        for i in range(len(x)):
            plt.text(i, y[i] + 10, y[i], ha = 'center')
    
    dataframe = dataframe.dropna()
    
    seating_type_counts = dataframe.groupby('seating_type').name.count().reset_index()
    seating = seating_type_counts.seating_type.values
    counts = seating_type_counts.name.values
    
    plt.figure(figsize = (10, 8))
    ax = plt.subplot()
    
    plt.bar(range(len(seating)), counts)
    ax.set_xticks(range(len(seating)))
    ax.set_xticklabels(seating, rotation = 90)
    
    plt.title("Seating Types on Roller Coasters")
    plt.xlabel("Seating Types")
    plt.ylabel("Count")
    addlabels(seating, counts)
    
    plt.show()
    
seating_type(roller_coasters)
```
