```python
import pandas as pd
import numpy as np
import re

pd.set_option('display.max_colwidth', None)

df = pd.read_csv('jeopardy.csv', skipinitialspace=True)
display(df.head(10))
```

### 1. Find rows that filters the dataset for questions that contains the words "King"* or "England".

*This will display all questions that have the word 'King', 'Kingdom', or any other word that contains 'King'; as well as 'England'.

```python
king_or_england = df[df.Question.str.contains('King | England')]
#.contains is case sensitive so it will only display 'King' or 'England'

display(king_or_england)
```

We are only looking for questions that contain just 'king' and 'England'.

```python
import warnings
warnings.filterwarnings('ignore', 'This pattern has match groups')

#only_king_or_england = df[df.Question.astype(str).str.contains('King | England', case=False)]
#This almost worked, but words like 'speaking', 'peeking', etc. would be selected.

#includeKeywords = [" King ", " England "]
#only_king_or_england = df[df.Question.str.contains('|'.join(includeKeywords), case=False).any(level=0)]
#This was closer, but it would only show 'king' and 'england' that had spaces in front and behind the word.

only_king_or_england = df[df.Question.str.contains(r"\b(King|England)\b", case=False)]

display(only_king_or_england)
```

### 3. Find the average of "Value" from the rows that contain 'King' or 'England'.

```python
df['Float of Value'] = df.Value #this copies the value to the new column
df['Float of Value'] = df.Value.astype(str).str.lstrip('$') #strips the '$'
df['Float of Value'] = df['Float of Value'].replace(',', '', regex=True) #gets rid of the ','
df['Float of Value'] = df['Float of Value'].replace("None", "0") #replaces "None" with '0'
df['Float of Value'] = pd.to_numeric(df["Float of Value"]) #turns the numbers(strings) into floats

only_king_or_england = df[df.Question.str.contains(r"\b(King|England)\b", case=False)]
only_king_or_england_value = only_king_or_england['Float of Value'].mean()
display(only_king_or_england)
print("The average of the values that contain either 'king' or 'England' is $" + str(round(only_king_or_england_value, 2)) + ".")
```

### 4. Find the number of unique answers from the dataset that contains only 'king' or 'England'.

```python
unique_answers = only_king_or_england['Answer'].value_counts()

print(unique_answers)
```

### 5. How many questions from the 1990s use the word (or words related to) "Computer" compared to questions from the 2000s?

```python
df['Year of Episode'] = pd.DatetimeIndex(df['Air Date']).year

year_1990 = df[df['Year of Episode'] == 1990]
year_1990_computer = year_1990[year_1990.Question.str.contains('Computer', case=False)]
display(year_1990_computer)
print("There are " + str(len(year_1990_computer)) + " questions that contain 'computer' in the 1990s.")

year_2000 = df[df['Year of Episode'] == 2000]
year_2000_computer = year_2000[year_2000.Question.str.contains('Computer', case=False)]
display(year_2000_computer)
print("There are " + str(len(year_2000_computer)) + " questions that contain 'computer' in the 2000s.")
```

### 6. Is there a connection between the 'Round' and the 'Category'? Are you more likely to find the 'Technology' category(ies) in Single Jeopardy or Double Jeopardy?

```python
single_jeopardy = df[df['Round'] == 'Jeopardy!']
single_jeopardy_category = single_jeopardy[single_jeopardy.Category.str.contains(r"\b(Technology)\b", case=False)]
display(single_jeopardy_category)

double_jeopardy = df[df['Round'] == 'Double Jeopardy!']
double_jeopardy_category = double_jeopardy[double_jeopardy.Category.str.contains(r"\b(Technology)\b", case=False)]
display(double_jeopardy_category)
```

You're more likely to find Technology categories in 'Jeopardy!' (also known as 'Single Jeopardy'), rather than 'Double Jeopardy!'.
