## Page Visits Funnel

Cool T-Shirts Inc. has asked you to analyze data on visits to their website. Your job is to build a funnel, which is a description of how many people continue to the next step of a multi-step process.

<br>

Our funnel is going to describe the following process:
1. A user visits CoolTShirts.com
2. A user adds a t-shirt to their cart
3. A user clicks “checkout”
4. A user actually purchases a t-shirt

<hr>

### Funnel for Cool T-Shirts Inc.

Inspect the DataFrames:

```python
import pandas as pd

visits = pd.read_csv('visits.csv',
                     parse_dates=[1])
cart = pd.read_csv('cart.csv',
                   parse_dates=[1])
checkout = pd.read_csv('checkout.csv',
                       parse_dates=[1])
purchase = pd.read_csv('purchase.csv',
                       parse_dates=[1])
```

```python
display(visits)
```

```python
display(cart)
```

```python
display(checkout)
```

```python
display(purchase)
```

Combine `visits` and `cart` using a *left merge*.

```python
visits_cart = pd.merge(visits, cart, how='left')
display(visits_cart)
```

```python
print(len(visits_cart))
```

How many of the timestamps are `null` for the column `cart_time`?

```python
visits_cart_null = visits_cart[visits_cart.cart_time.isnull()]
display(visits_cart_null)
```

What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?

```python
not_place_order_visits_cart = (float(len(visits_cart_null)) / float(len(visits_cart.visit_time.notnull()))) * 100
print(str(not_place_order_visits_cart) + "% of users visited the website, but did not place an item in their cart.")
```

Repeat the left merge for `cart` and `checkout` and count `null` values. What percentage of users put items in their cart, but did not proceed to checkout?

```python
cart_checkout = pd.merge(cart, checkout, how='left')
display(cart_checkout)
```

```python
cart_checkout_null = cart_checkout[cart_checkout.checkout_time.isnull()]
display(cart_checkout_null)
```

```python
not_place_order_cart_checkout = (float(len(cart_checkout_null)) / float(len(cart_checkout.cart_time.notnull()))) * 100
print(str(not_place_order_cart_checkout) + "% of users put items in their cart, but did not checkout.")
```

Merge all four steps of the funnel, in order, using a series of left merges.

```python
all_data = visits.merge(cart, how='left').merge(checkout, how='left').merge(purchase, how='left')
display(all_data.head())
```

What percentage of users proceeded to checkout, but did not purchase a t-shirt?

```python
purchase_null = all_data[all_data.purchase_time.isnull()]
display(purchase_null)
```

```python
not_purchase = (float(len(purchase_null)) / float(len(all_data.checkout_time.notnull()))) * 100
print(str(not_purchase) + "% of users proceeded to checkout, but did not purchase a t-shirt. \n\nTherefore, we can assume that users who visit the website are more than likely to not purchase a t-shirt.")
```

### Average Time to Purchase

Let’s calculate the average time from initial visit to final purchase.

```python
all_data['time_to_purchase'] = all_data.purchase_time - all_data.visit_time
display(all_data)
```

Calculate the average time to purchase.

```python
print(all_data.time_to_purchase.mean())
print("It took an average of 43 minutes and 53 seconds from visiting the website, to purchasing a t-shirt.")
```
